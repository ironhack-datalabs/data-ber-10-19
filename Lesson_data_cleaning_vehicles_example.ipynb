```python
df["barrels08"].head()
```

```python
# every 90% of the cars are below this point: 23.54...

np.percentile(df["barrels08"], 90)
```

```python
def detect_low_variance(df, column, perc):
    min_value = min(df[column])
    perc_value = np.percentile(df[column], perc)
    return min_value == perc_value
```

```python
detect_low_variance(df, "barrels08", 90)
```

```python
# only give me the columns with numeric data type
# inlude takes a list, therefore []
# .columns to get a list only of column names

df.select_dtypes(include=[np.number]).columns
```

```python
# function to display columns with low variance in a list.
# percentile (90) can be changed
    
low_variance = [col for col in df.select_dtypes(include=[np.number]).columns if detect_low_variance(df, col, 90)]
```

```python
df = df.drop(low_variance, axis=1)
```

# 3. Outliers
Using IQR (Interquartile Range)

```python
# transpose to change columns to rows "Transponieren"

df_desc = df.describe().transpose()
```

```python
# assign to create new column

df_desc = df_desc.assign(IQR=df_desc["75%"]-df_desc["25%"])
```

```python
df_desc
```

```python
df_desc.index
```

```python
# create empty data frame with already existing columns of Data Set used before

outliers = pd.DataFrame(columns=df.columns)

# iterating over index:

for col in df_desc.index:
    iqr = df_desc.loc[col, "IQR"]
    cutoff = iqr * 6
    lower_threshold = df_desc.loc[col, "25%"] - cutoff
    upper_threshold = df_desc.loc[col, "75%"] + cutoff
    results = df[(df[col]< lower_threshold) | (df[col] > upper_threshold)].copy()
    results = results.assign(outlier=col)
    outliers = pd.concat([outliers, results], ignore_index = True, sort = False)
```

```python
# groupby does not sort data

outliers.groupby("outlier")["outlier"].count()
```

```python
# value_counts sorts data desc by default value

outliers["outlier"].value_counts()
```
